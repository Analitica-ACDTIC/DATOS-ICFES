```python
df_icfes_all = df_icfes_all.loc[df_icfes_all['FAMI_ESTRATOVIVIENDA']!='-']
```

```python
def cast_dedicacion_internet(x):
    if x!=x:
        return 'NR'
    else:
        if x=='No Navega Internet':
            return 'NO'
        elif x=='-':
            return 'NR'
        else:
            return 'SI'
```

```python
df_icfes_all['ESTU_DEDICACIONINTERNET'] = df_icfes_all['ESTU_DEDICACIONINTERNET'].apply(cast_dedicacion_internet)
```

```python
df_icfes_all = df_icfes_all.loc[df_icfes_all['ESTU_DEDICACIONINTERNET']!='NR']
```

```python
df_icfes_all['FAMI_ESTRATOVIVIENDA'].value_counts()
```

```python
df_icfes_all = df_icfes_all.loc[df_icfes_all['FAMI_ESTRATOVIVIENDA'].isin([
    'Sin Estrato',
    'Estrato 1',
    'Estrato 2',
    'Estrato 3'
    ])]
```

```python
df_icfes_all['Estrato - Internet'] = df_icfes_all['FAMI_ESTRATOVIVIENDA'] + ' - ' + df_icfes_all['ESTU_DEDICACIONINTERNET'] 
```

```python
df_icfes_all['ESTU_DEPTO_RESIDE'] = df_icfes_all['ESTU_DEPTO_RESIDE'].apply(
    lambda x:x if x!='BOGOTÁ' else 'BOGOTA'
)
```

```python
df_agg = df_icfes_all.groupby([
    'PERIODO','ESTU_DEPTO_RESIDE','FAMI_ESTRATOVIVIENDA','ESTU_DEDICACIONINTERNET','Estrato - Internet'
]).agg({
    'PUNT_GLOBAL':'mean',
    'ESTU_CONSECUTIVO':'count'
}).reset_index().rename(
    columns={
        'ESTU_CONSECUTIVO':'Conteo'
    }
)
```
